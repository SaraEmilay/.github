```python
from geopy.distance import great_circle

def plot_scatter(df, title):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='Consumo', y=df.index,
                    hue='Fraude',
                    palette={0: 'blue', 1: 'red'},
                    alpha=0.7,
                    style='Fraude',
                    markers={0: 'o', 1: 'X'})
    plt.title(title)
    plt.xlabel('Consumo')
    plt.ylabel('Índice (cada ponto representa uma entrada)')
    plt.axhline(0, color='gray', lw=1, ls='--')
    plt.legend(title='Fraude', loc='upper right', labels=['Não Fraude', 'Fraude'])
    plt.grid()
    plt.show()

def encontrar_imoveis_no_raio(df, radius=500):

    imoveis_no_raio = pd.DataFrame(columns=['IMOV_ID', 'Fraude_ID'])

    for idx, row in df[df['Fraude'] == 1].iterrows():
        fraudes_loc = (row['latitude'], row['longitude'])

        for idx2, row2 in df.iterrows():
            imovel_loc = (row2['latitude'], row2['longitude'])
            distancia = great_circle(fraudes_loc, imovel_loc).meters

            if distancia <= radius:
                imoveis_no_raio = imoveis_no_raio.append({'IMOV_ID': row2['IMOV_ID'], 'Fraude_ID': row['IMOV_ID']}, ignore_index=True)

    return imoveis_no_raio
```

```python
import pandas as pd
from geopy.distance import great_circle

dataframes_selecionados = {
    'df_2019_7': df_2019_7,
    'df_2020_1': df_2020_1,
    'df_2020_12': df_2020_12,
    'df_2021_1': df_2021_1,
    'df_2021_11': df_2021_11
}

def calcular_pontos_no_raio(df, radius=500):
    pontos_no_raio = []
    fraudes = df[df['Fraude'] == 1]

    for idx, row in fraudes.iterrows():
        fraude_loc = (row['latitude'], row['longitude'])
        for idx2, row2 in df.iterrows():
            imovel_loc = (row2['latitude'], row2['longitude'])
            if pd.notna(fraude_loc[0]) and pd.notna(fraude_loc[1]) and pd.notna(imovel_loc[0]) and pd.notna(imovel_loc[1]):
                distancia = great_circle(fraude_loc, imovel_loc).meters

                if distancia <= radius:
                    pontos_no_raio.append(row2)

    return pd.DataFrame(pontos_no_raio)

resultados_dentro_do_raio = {}

for nome, df in dataframes_selecionados.items():
    df = df.dropna(subset=['latitude', 'longitude'])

    pontos_no_raio = calcular_pontos_no_raio(df)

    if not pontos_no_raio.empty:
        resultados_dentro_do_raio[nome] = pontos_no_raio

caminho_base = '/content/drive/MyDrive/ColabPIBIC/clientes_proximos_para_o_teste_de_hipotese/'

for nome, df_resultado in resultados_dentro_do_raio.items():
    nome_arquivo = f'{caminho_base}{nome}_dentro_do_raio.csv'
    df_resultado.to_csv(nome_arquivo, index=False)
    print(f"Arquivo salvo: {nome_arquivo} com {df_resultado.shape[0]} linhas dentro do raio.")

df_final = pd.concat(resultados_dentro_do_raio.values(), keys=resultados_dentro_do_raio.keys())
df_final.to_csv(f'{caminho_base}todos_dentro_do_raio.csv', index=False)
print(f"Arquivo salvo: {caminho_base}todos_dentro_do_raio.csv com {df_final.shape[0]} linhas.")
```

```python
import statsmodels.api as sm

for nome, df in dataframes_importados.items():
    media_nao_fraude = medias_consumo[nome]

    df['Consumo_acima_media'] = (df['Consumo'] > media_nao_fraude).astype(int)

    X = df['Consumo']
    y = df['Consumo_acima_media']

    X = sm.add_constant(X)

    modelo = sm.Logit(y, X).fit()

    print(f'\nResumo do modelo de regressão logística para {nome}:')
    print(modelo.summary())
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve


# Separando as variáveis independentes (X) e a dependente (y)
X = df_resampled.drop(columns=['Fraude'])  # Todas as colunas menos a variável alvo
y = df_resampled['Fraude']  # Variável alvo: Fraude

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando e treinando o modelo de regressão logística
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Fazendo previsões nos dados de teste
y_pred = log_reg.predict(X_test)

# Avaliando a acurácia
print("Acurácia:", accuracy_score(y_test, y_pred))

# Exibindo o relatório de classificação (precision, recall, f1-score)
print("Relatório de Classificação:\n", classification_report(y_test, y_pred))

# Matriz de Confusão
print("Matriz de Confusão:\n", confusion_matrix(y_test, y_pred))

# Calculando o AUC-ROC
y_pred_prob = log_reg.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_prob)
print("AUC-ROC:", roc_auc)

# Plotando a curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
